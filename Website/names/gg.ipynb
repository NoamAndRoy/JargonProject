```python
import json, csv
```

```python
with open('./firstnames.json') as js:
    a = json.load(js)
aa = set(filter(lambda x: len(x) > 1, a.keys()))
aa
```

```python
with open('./first-names.txt') as txt:
    b = txt.readlines()

bb = set(filter(lambda x: len(x) > 1, map(lambda y: y[:-1], b)))
bb
```

```python
nana = sorted(aa|bb)
nana
```

```python
with open('names.csv', 'w+', newline='') as names:
    csv_writer = csv.writer(names)

    for n in nana:
        csv_writer.writerow([n])
```

```python
with open('names.csv', 'r') as ns:
    kkk = csv.reader(ns)
    cc = [r[0] for r in kkk]

cc
```
